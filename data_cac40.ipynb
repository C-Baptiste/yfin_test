```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
list_cac40 = ['AI.PA', 'AIR.PA', 'ALO.PA', 'MT.AS', 'CS.PA', 'BNP.PA', 'EN.PA', 'CAP.PA', 'CA.PA', 'ACA.PA', 
              'BN.PA', 'DSY.PA', 'ENGI.PA', 'EL.PA', 'ERF.PA', 'RMS.PA', 'KER.PA', 'OR.PA', 'LR.PA', 'MC.PA', 
              'ML.PA', 'ORA.PA', 'RI.PA', 'PUB.PA', 'RNO.PA', 'SAF.PA', 'SGO.PA', 'SAN.PA', 'SU.PA', 'GLE.PA', 
              'STLA.PA', 'STM.PA', 'TEP.PA', 'HO.PA', 'TTE.PA', 'URW.AS', 'VIE.PA', 'DG.PA', 'VIV.PA', 'WLN.PA']
```

```python
datas = yf.Ticker(list_cac40[20]) #.info.get('priceToBook')

print(datas.info.get('symbol'))
```

```python
list_shortName = []
list_sector = []
list_priceToBook = []
list_dividendYield = []
list_pegRatio = []
list_beta = []
list_ebitdaMargins = []
```

```python
print(list_priceToBook)
```

```python
d = {
        "symbol": list_cac40,
        "shortName": list_shortName,
        "sector": list_sector,
        "priceToBook": list_priceToBook,
        "dividendYield": list_dividendYield,
        "pegRatio": list_pegRatio,
        "beta": list_beta,
        "ebitdaMargins": list_ebitdaMargins
}
```

```python
for company in list_cac40:
    data = yf.Ticker(company)
    list_shortName.append(data.info.get('shortName'))
    list_sector.append(data.info.get('sector'))
    list_priceToBook.append(data.info.get('priceToBook'))
    print(data.info.get('symbol'))
    print(data.info.get('shortName'))
    print(data.info.get('priceToBook'))
    list_dividendYield.append(data.info.get('dividendYield'))
    list_pegRatio.append(data.info.get('pegRatio'))
    list_beta.append(data.info.get('beta'))
    list_ebitdaMargins.append(data.info.get('ebitdaMargins'))
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
df.describe()
```

```python
df.sort_values(by="priceToBook", inplace=True)
plt.subplots(figsize=(10, 10))
plt.barh(df["shortName"], df["priceToBook"])
plt.show()
```

```python
df.sort_values(by="pegRatio", inplace=True)
plt.subplots(figsize=(10, 5))
plt.barh(df["shortName"], df["pegRatio"])
plt.show()
```

```python
df.sort_values(by="dividendYield", inplace=True)
plt.subplots(figsize=(10, 10))
plt.barh(df["shortName"], df["dividendYield"])
plt.show()
```

```python
df.sort_values(by="ebitdaMargins", inplace=True)
plt.subplots(figsize=(10, 10))
plt.barh(df["shortName"], df["ebitdaMargins"])
plt.show()
```

```python
df.sort_values(by="beta", inplace=True)
plt.subplots(figsize=(10, 10))
plt.barh(df["shortName"], df["beta"])
plt.show()
```
