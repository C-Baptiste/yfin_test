```python
yf.Ticker('AI.PA').info
```

```python
list_shortName = []
list_priceToBook = []
list_dividendRate = []
```

```python
d = {
        "symbol": list_cac40,
        "shortName": list_shortName,
        "priceToBook": list_priceToBook,
        "dividendRate": list_dividendRate
}
```

```python
for company in list_cac40:
    data = yf.Ticker(company)
    list_shortName.append(data.info.get('shortName'))
    list_priceToBook.append(data.info.get('priceToBook'))
    list_dividendRate.append(data.info.get('dividendRate'))    
```

```python
df = pd.DataFrame(d)
```
